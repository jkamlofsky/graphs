# Trees

## 1. Basic Trees:

### 1.1. Libraries

```python
# Libraries
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
import time
```

### 1.2. Balanced m-ary  Tree

An m-ary Tree is considered "balanced" when all levels except the last one are integrated at the maximum capacity of nodes.

Next cell shows you how to draw a binary balanced tree of a given height. The parameters are: the height and the value of "m".

```python
m=2
level=3
T = nx.balanced_tree(m, level)
pos = graphviz_layout(T, prog="dot")
nx.draw(T, pos)
plt.show()
```

## 2. Covering trees

### 2.1. Getting a graph from a csv file

```python
df = pd.read_csv('files/airports_arg_edges.csv',sep=';')
G = nx.from_pandas_edgelist(df, source='source', target='target', edge_attr='name', create_using=nx.Graph())
pos = nx.shell_layout(G)
nx.draw(G,pos,with_labels = True)
nx.draw_networkx_edge_labels(G,pos,edge_labels=nx.get_edge_attributes(G,'name'))
```

### 2.2. Deep-First-Search (dfs) Tree:

```python
# DFS Tree Definition
dfst=nx.dfs_tree(G, 'AEP')
# Showing the Tree
pos = graphviz_layout(dfst, prog="dot")
nx.draw(dfst,pos,with_labels = True)
```

### 2.3. Breadth-First-Search (dfs) Tree:

```python
# BFS Tree Definition
bfst=nx.bfs_tree(G, 'AEP')
# Showing the Tree
pos = graphviz_layout(bfst, prog="dot")
nx.draw(bfst,pos,with_labels = True)
```
